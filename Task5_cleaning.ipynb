#3. Handle numeric and categorial missing values appropriately


Numeric columns are filled using mean()

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
```

Categorical columns are filled using mode()

```python
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
```

Columns with excessive missing data are dropped.

```python
df.drop(columns=['Cabin'], inplace=True, errors='ignore')
```

#4. A new column "Age_Category" is created using business logic to classify passengers into Child, Adult, and Senior groups.

```python
def age_category(age):
  if age < 18:
    return "Child"
  elif age < 60:
    return "Adult"
  else:
    return "Senior"

df['Age_Category'] = df['Age'].apply(age_category)
```
